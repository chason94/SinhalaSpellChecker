```python
def get_words(filename):
    with open(filename, 'r+', encoding='utf8') as f:
      return [x.strip() for x in f.read().split() if not bool(re.match(r'[A-z0-9]',x))]
```

```python
import re
rep_dict = {}
replacements = 'dataset/train/train_cleaning/replace.txt'
with open(replacements, 'r+', encoding = 'utf8') as file:
    repl_lst = file.readlines()
    for line in repl_lst:
        word = re.findall('[\u0D80-\u0DFF\u200d\u200c\u200b]{1,}', line)
        sign = re.findall('[<>]', line)
        # print(sign)
        if len(sign) == 0  or len(sign) > 1 :
            continue 
        if sign[0] == '>':
            rep_dict[word[0]] = word[1] 
        elif sign[0] == '<':
            rep_dict[word[1]] = word[0]
        else:
            pass 
deletions = 'dataset/train/train_cleaning/delete.txt'
with open(replacements, 'r+', encoding = 'utf8') as file:
    repl_lst = file.readlines()
    for line in repl_lst:
        word = re.findall('[\u0D80-\u0DFF\u200d\u200c\u200b]{1,}', line)
        rep_dict[word[0]] = '<unk>'
# rep_dict
```

```python
spell_corrected_1 = []
for i, word in enumerate(spell_corrected):
    correct_word = rep_dict.get(word)
    if correct_word:
        # print(word, '=>', correct_word)
        spell_corrected_1.append(correct_word)
    else:
        spell_corrected_1.append(word)
with open('output.txt', 'w+', encoding='utf8') as f:
    for word in spell_corrected_1:
        f.write(' {}'.format(word))
        if word == '<eos>':
            f.write('\n')
len(set(spell_corrected_1))
```
